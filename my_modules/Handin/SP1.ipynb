```python
import pandas as pd
import folium
data = pd.read_csv("https://wfs-kbhkort.kk.dk/k101/ows?service=WFS&version=1.0.0&request=GetFeature&typeName=k101:monumenter&outputFormat=csv&SRSNAME=EPSG:4326")
print(data)
#1 Hvor mange monumenter er der i København?
print(data.shape)
#2 Hvor mange monumenter bliver vedligeholdt? Dvs. graffitirenhold = ja
data_graffiti = data.loc[data['graffitirenhold']=='Ja']
print("Der er",len(data_graffiti),"monumenter som er blevet vedligeholdt")
#3 Lav en funktion som kan finde koordinaterne på et monument baseret på monumentets id eller navn?
def monumentById(monumentId):
    monumentId = data.loc[data['id']==monumentId]
    longitude = monumentId.iloc[0,7]
    latitude = monumentId.iloc[0,8]
    return (latitude, longitude)
pos = monumentById(50063.0)
print(monumentById(50063.0))

#3.a Vis monument som bliver returneret i metoden, på et kort over københavn ved brug af plotting.
map_osm = folium.Map(location=(pos), zoom_start=10) # here we draw the map.
folium.Marker(pos).add_to(map_osm)         # here we plot the environmental zone 1
map_osm.save('./osm.html')
map_osm
#5 Lav en metode der optegner alle monumenterne på kortet ved brug af plotting.
map_osm = folium.Map(location=(pos), zoom_start=10)
for x in range(len(data)):
    folium.Marker((data.iloc[x, 8], data.iloc[x, 7])).add_to(map_osm)
map_osm.save('./osm.html')
map_osm
    
```
