 1. Reducer datasettet til Honda'er med benzinmotor
 2. Lav linear regression på hestekræfter og highway MPG hvor x er hestekræfter og y er MPG
 3. Hvad er koefficienten?
 4. Tegn dataen i et scatterplot med regressions ligningen
 5. Hvad vil MPG være når en motor har 200 hestekræfter?

```python
import pandas as pd

#1
df = pd.read_csv('https://think.cs.vt.edu/corgis/datasets/csv/cars/cars.csv')

df = df.loc[(df['Identification.Make'] == 'Honda') & (df['Fuel Information.Fuel Type'])]
df
```

```python
import numpy as np
import sklearn.linear_model
#2
xs = df['Engine Information.Engine Statistics.Horsepower']
ys = df['Fuel Information.Highway mpg']
xs_reshape = np.array(xs).reshape(-1, 1)

model = sklearn.linear_model.LinearRegression()
print(model.fit(xs_reshape, ys))

#3
print(model.coef_)
print(model.intercept_)
```

```python
import matplotlib.pyplot as plt
#4
plt.scatter(xs, ys)

plt.plot(xs, model.coef_*xs+model.intercept_)
```

```python
#5
predicted = model.predict(xs_reshape)
hestekræfter = model.predict([[200]])

print('MPG vil være på: ', hestekræfter[0])
```
